```python
# filtering to get recommandation from specific collection
def filtering(data,collection):
  if collection is None:
    return data.drop(columns=['belongs_to_collection','id'],axis=1)
  else:
    data=data[(data['belongs_to_collection']==collection)]
    return data.drop(['belongs_to_collection','id'],axis=1)

o=filtering(r_data,'James Bond Collection')
o.shape
```
